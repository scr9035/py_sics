#Determine the $B_{ig}$ symmetry channel: Slope subtraction method

1. Input two files: $\frac{dR_{yy}}{d\epsilon_{ii}}$, $\frac{dR_{xx}}{d\epsilon_{ii}}$
2. Set the value for the piezo Poisson ratio
3. Set the strain direction used to calculate the slop 'eyy' or 'exx'
4. Determine interoplating range and step size
5. Save file

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

##1. Input $\frac{dR_{yy}}{d\epsilon_{ii}}$, $\frac{dR_{xx}}{d\epsilon_{ii}}$ files

```python
file_Ryy = np.genfromtxt('URu2Si2_dRyy_exx.txt')
file_Rxx = np.genfromtxt("URu2Si2_dRyy_eyy.txt")

#sort increasing x, need this for the fitting routine to work properly
file_Ryy = file_Ryy[np.argsort(file_Ryy[:,0])]
file_Rxx = file_Rxx[np.argsort(file_Rxx[:,0])]
```

##2. Set the poisson_ratio value. 
##3. Set the strain direction used to calculate the slope ('eyy' or 'exx')
##4. Set interoplation range. np.arange(X,Y,Z) interoplate from X to Y in steps of Z 

```python
poisson_ratio = 1.2 #depends on the strain gauge
strain_type = 'eyy' # must be either 'exx' or 'eyy'


if strain_type == 'eyy':
    strain_direction = (1+poisson_ratio)
else:
    strain_direcction = -1*poisson_ratio
    

#interpolate eyy and exx such that they have 1 point per x Kelvin steps

interpX = np.arange(10,65,0.1) #make from 2 to 55 in steps of .1 var's
interpY_file_Ryy = np.interp(interpX, file_Ryy[:,0], file_Ryy[:,1])
interpY_file_Rxx = np.interp(interpX, file_Rxx[:,0], file_Rxx[:,1])

symmetry_poisson = interpY_file_Rxx - interpY_file_Ryy



pureSymChannel = symmetry_poisson / (strain_direction)
#do the poisson math


plt.figure()
plt.ylabel(r'$\frac{dR_{ii}}{d\epsilon_{ii}}$')
plt.xlabel('Temperature')
plt.plot(file_Ryy[:,0], file_Ryy[:,1],
         file_Rxx[:,0], file_Rxx[:,1],
         interpX, interpY_file_Rxx, 
         interpX, interpY_file_Ryy)
```

```python
#Plot just the 2m66 data

plt.figure()
plt.ylabel(r'$2m_{66}$')
plt.xlabel('Temperature')
plt.plot(interpX, symmetry_poisson,
         interpX, pureSymChannel)
```

##3. Input file name to save. 
###Saves two columns, tuning parameter (field or temp) and $B_{ig}$

```python
#output save file txt file with data wanted
saveSymCh = np.asarray([interpX,pureSymChannel])
saveFileName = "testSave.txt"
np.savetxt(saveFileName, np.transpose(saveSymCh), delimiter=" ")
```
